```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("metadata_h3nx_ha.tsv", sep="\t", lineterminator="\n")
```

```python
#subtype plotting, how many sequences per subtype

# print(df.groupby('subtype').size())
subtype_counts = df["subtype"].value_counts()

# create a bar plot using the above counts
ax = subtype_counts.plot(kind="bar")

ax.set_xlabel("Subtype")
ax.set_ylabel("# of sequences")

for container in ax.containers:
    ax.bar_label(container, label_type='edge', fontsize=10)

# show the plot
plt.show()
```

```python
#sequences per region

region_counts = df["region"].value_counts()
ax = region_counts.plot(kind="bar")

ax.set_xlabel("region")
ax.set_ylabel("# of sequences")

for container in ax.containers:
    ax.bar_label(container, label_type='edge', fontsize=10)

# # show the plot
plt.show()
```

```python
#sequences per country

country_counts = df["country"].value_counts()
print(country_counts[4:].sum())
ax1 = country_counts[:10].plot(kind="bar")

# set the x-axis label
ax1.set_xlabel("Country")
# set the y-axis label
ax1.set_ylabel("# of sequences")

for container in ax1.containers:
    ax1.bar_label(container, label_type='edge', fontsize=8)

fig = plt.gcf()
fig.set_size_inches(8, 4)

# show the plot
plt.show()
```

```python
grouped_data = df.groupby('host').size()


graph1_data = grouped_data[grouped_data.index.isin(['Avian', 'Canine', 'Swine', 'Equine'])]


fig, (ax1,ax2) = plt.subplots(2, 1, figsize = (10,15))
graph1_data.plot(kind='bar', ax=ax1)

ax1.set_title('Number of sequences per host')
ax1.set_xlabel('Host')
ax1.set_ylabel('# of sequences')

ax1.xaxis.set_label_coords(0.5, -0.05)

graph2_data = grouped_data[~grouped_data.index.isin(['Avian', 'Canine', 'Swine', 'Equine'])]


graph2_data.plot(kind='bar', ax=ax2)


ax2.set_xlabel('Host')
ax2.set_ylabel('# of sequences')

# #add text labels for each bar
for ax in fig.axes:
    for container in ax.containers:
        ax.bar_label(container, label_type='edge', fontsize=10)

ax2.xaxis.set_label_coords(0.5, -0.05)
ax2.set_xlabel(xlabel = None)

# Adjust the spacing between the subplots
plt.subplots_adjust(hspace=0.7)

# Show the plot
plt.show()
```

```python
#host per country

hosts_of_interest = ['Feline', 'Canine', 'Swine', 'Equine', 'Avian']
for host in hosts_of_interest:
    #filter the dataframe to only include the current host
    df_filtered = df[df['host'] == host]

    #group the filtered dataframe by country and count the occurrences
    df_grouped = df_filtered.groupby('country').size().reset_index(name='# of sequences')

    #plot a bar graph of the current host and its countries
    ax = df_grouped.plot(x='country', y='# of sequences', kind='bar', figsize=(10, 6), width=0.8)

   
    #add labels to each bar
    for i, count in enumerate(df_grouped['# of sequences']):
        ax.text(i, count + 1, str(count), ha='center', va='bottom')

    # add axis labels and a title
    ax.set_xlabel('Country')
    ax.set_ylabel('# of sequences')
    ax.set_title(f'{host} Virus Sequences by Country')

    # show the plot
    plt.show()
```
